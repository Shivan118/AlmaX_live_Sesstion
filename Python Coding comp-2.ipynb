````python
def calculate_depths(input_string):
    # Step 1: Split the input string into separate groups
    groups = input_string.split()

    # Step 2: Initialize a list to store the depth of each group
    depths = []

    # Step 3: Process each group to calculate its depth
    for group in groups:
        current_depth = 0
        max_depth = 0

        # Traverse each character in the group ->  "(()) (()) () ((()()()))"
        for char in group:
            if char == '(':
                current_depth += 1
                # Update max depth if current depth is greater
                max_depth = max(max_depth, current_depth)
            elif char == ')':
                current_depth -= 1

        # Append the max depth of the current group to the depths list
        depths.append(max_depth)

    # Step 4: Return the list of depths
    return depths

# Test input
input_string = "(()) (()) () ((()()()))"

# Get the output
output = calculate_depths(input_string)

# Print the output
print(output)
````

### <span style="color:green;"> Q3) Remove All Adjacent Duplicates In String </span>

You are given a string s consisting of lowercase English letters. A duplicate removal consists of choosing two adjacent and equal letters and removing them.

We repeatedly make duplicate removals on s until we no longer can.

Return the final string after all such duplicate removals have been made. It can be proven that the answer is unique.



**Input:** s = abbaca

**Output:** ca

**Explanation:**
For example, in "abbaca" we could remove "bb" since the letters are adjacent and equal, and this is the only possible move.  The result of this move is that the string is "aaca", of which only "aa" is possible, so the final string is "ca".

#### <span style="color:green">Solution:</span>

#### Steps to Solve the Problem

1. **Initialize an Empty Stack:** This will be used to keep track of characters as we process the string.

2. **Iterate Through Each Character:** For each character in the string, check if it matches the top character of the stack.
   - **If it Matches:** This means we have found two adjacent duplicates. Pop the top character from the stack (remove the duplicate pair).
   - **If it Does Not Match:** Push the current character onto the stack.

3. **Build the Resulting String:** After processing all characters, the stack will contain the final string with all adjacent duplicates removed. Convert this stack into a string.

#### Example

For the input string `s = "abbaca"`:

- **Start with an empty stack:** `[]`

- **Process each character:**
  - `'a'`: Stack becomes `['a']`
  - `'b'`: Stack becomes `['a', 'b']`
  - `'b'`: Stack becomes `['a']` (since `'b'` was removed)
  - `'a'`: Stack becomes `['a', 'a']`
  - `'c'`: Stack becomes `['a', 'c']` (since `'a'` was removed)
  - `'a'`: Stack becomes `['c', 'a']`

- **Final stack:** `['c', 'a']`


````python
def remove_adjacent_duplicates(s: str) -> str:
    stack = []
    for char in s:  # a,b,b,a,c,a
        if stack and stack[-1] == char:
            stack.pop()  # Remove the top of the stack if it's the same as the current character
        else:
            stack.append(char)  # Add the current character to the stack
    return ''.join(stack)  # Convert the stack to a string

# Example usage
s = "abbaca"
print(remove_adjacent_duplicates(s))  # Output: "ca"
````

   - Use `sorted(words, key=lambda w: int(w[-1]))` to sort the words where the key is the number extracted from the end of each word.
   - **Output**: A sorted list of words in the correct order based on the numerical position, e.g., `['Me1', 'Myself2', 'and3', 'I4']`.

3. **Reconstruct the Sentence Without Position Numbers**:
   - Initialize an empty string `result` to build the final sentence.
   - Iterate through the sorted list of words and remove the numerical position by slicing off the last character from each word.

     - After appending each word, print the intermediate state of `result` to show how the sentence is being reconstructed:

#### <span style="color:green">Solution:</span>

### Steps to Merge Two Strings Alternately

1. **Initialize Variables**:
   - Create an empty list to build the merged string.
   - Use two pointers (or indices) to keep track of the current position in `word1` and `word2`.

2. **Iterate Over the Strings**:
   - Use a loop to alternate between characters of `word1` and `word2` until one of the strings is exhausted.
   - Append characters from both strings to the merged list in alternating order.

3. **Append Remaining Characters**:
   - After the loop, append any remaining characters from the longer string to the merged list.

4. **Convert List to String**:
   - Join the list of characters into a single string.

````python
def mergeAlternately(word1, word2):
    # Initialize pointers and result list
    i, j = 0, 0
    merged = []

    # Merge characters from both words alternately
    while i < len(word1) and j < len(word2):
        merged.append(word1[i])
        merged.append(word2[j])
        i += 1
        j += 1
    
    # Append remaining characters from word1 if any
    while i < len(word1):
        merged.append(word1[i])
        i += 1
    
    # Append remaining characters from word2 if any
    while j < len(word2):
        merged.append(word2[j])
        j += 1
    
    # Convert list to string and return
    return ''.join(merged)

# Example usage
word1 = "abc"
word2 = "pqr"
print(mergeAlternately(word1, word2))  # Output: "apbqcr"
````

### Step-by-Step Execution of `mergeAlternately`

1. **Initialize Pointers and Result List**:
   - `i, j = 0, 0`: Pointers to track the current position in `word1` and `word2`, respectively.
   - `merged = []`: An empty list to build the merged string.

   - **Loop Condition**: `while i < len(word1) and j < len(word2)`: Continue merging characters as long as there are characters left in both strings.
     - **Action**: 
       - Append `word1[i]` to `merged`.
       - Append `word2[j]` to `merged`.
       - Increment `i` and `j` to move to the next character in each string.
   - **Example Execution**:
     - `word1 = "abc"`, `word2 = "pqr"`
     - Iteration 1: `i=0`, `j=0` -> `merged = ['a', 'p']`
     - Iteration 2: `i=1`, `j=1` -> `merged = ['a', 'p', 'b', 'q']`
     - Iteration 3: `i=2`, `j=2` -> `merged = ['a', 'p', 'b', 'q', 'c', 'r']`
     - Loop ends when either `i` or `j` reaches the end of the respective string.

3. **Append Remaining Characters from `word1`**:
   - **Condition**: `while i < len(word1)`: Check if there are any remaining characters in `word1`.
   - **Action**: Append remaining characters to `merged`.
   - **Example Execution**:
     - If `word1` had more characters, append them to `merged`.
     - In this example, `word1` has no remaining characters after the loop.

4. **Append Remaining Characters from `word2`**:
   - **Condition**: `while j < len(word2)`: Check if there are any remaining characters in `word2`.
   - **Action**: Append remaining characters to `merged`.
   - **Example Execution**:
     - If `word2` had more characters, append them to `merged`.
     - In this example, `word2` has no remaining characters after the loop.

5. **Convert List to String and Return**:
   - **Action**: Use `''.join(merged)` to convert the list of characters into a single string.
   - **Return**: The final merged string is returned.
   - **Example Execution**:
     - `merged = ['a', 'p', 'b', 'q', 'c', 'r']`
     - `''.join(merged)` -> `'apbqcr'`

````python
def mergeAlternately(word1, word2):
    # Initialize pointers and result list
    i, j = 0, 0
    merged = []
    print("Initial State: i =", i, ", j =", j, ", merged =", merged)

    # Merge characters from both words alternately
    while i < len(word1) and j < len(word2):
        merged.append(word1[i])
        merged.append(word2[j])
        print("Appending:", word1[i], "and", word2[j])
        print("Current merged list:", merged)
        i += 1
        j += 1
        print("Updated State: i =", i, ", j =", j, ", merged =", merged)

    # Append remaining characters from word1 if any
    while i < len(word1):
        merged.append(word1[i])
        print("Appending remaining from word1:", word1[i])
        print("Current merged list:", merged)
        i += 1
        print("Updated State: i =", i, ", merged =", merged)

    # Append remaining characters from word2 if any
    while j < len(word2):
        merged.append(word2[j])
        print("Appending remaining from word2:", word2[j])
        print("Current merged list:", merged)
        j += 1
        print("Updated State: j =", j, ", merged =", merged)

    # Convert list to string and return
    result = ''.join(merged)
    print("Final merged string:", result)
    return result

# Example usage
word1 = "abc"
word2 = "pqr"
print(mergeAlternately(word1, word2))  # Output: "apbqcr"
````
